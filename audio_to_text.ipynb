```python
import sounddevice as sd
import numpy as np
import soundfile as sf
import speech_recognition as sr
import threading
import keyboard  # pip install keyboard
import pyttsx3

fs = 44100  # Sampling rate
recording = []
is_recording = True

def record_callback(indata, frames, time, status):
    if is_recording:
        recording.append(indata.copy())

def start_recording():
    with sd.InputStream(samplerate=fs, channels=1, dtype='int16', callback=record_callback):
        print("üéôÔ∏è Recording... Press Enter to stop.")
        keyboard.wait('enter')

def main():
    global is_recording

    # Start recording thread
    thread = threading.Thread(target=start_recording)
    thread.start()
    thread.join()
    is_recording = False

    # Combine and save audio
    audio = np.concatenate(recording, axis=0)
    sf.write("temp.wav", audio, fs)

    # Ask user to select language
    print("Choose language for recognition:")
    print("1. Hindi")
    print("2. Telugu")
    print("3. Tamil")
    print("4. Kannada")
    print("5. English")
    choice = input("Enter choice number: ").strip()

    lang_codes = {
        "1": "hi-IN",
        "2": "te-IN",
        "3": "ta-IN",
        "4": "kn-IN",
        "5": "en-US"
    }
    selected_lang = lang_codes.get(choice, "en-US")

    # Recognize speech
    r = sr.Recognizer()
    with sr.AudioFile("temp.wav") as source:
        audio_data = r.record(source)

        try:
            text = r.recognize_google(audio_data, language=selected_lang)
            print(f"üó£Ô∏è You said ({selected_lang}): {text}")
        except sr.UnknownValueError:
            print("‚ùå Could not understand the audio.")
            return
        except sr.RequestError as e:
            print(f"‚ùå Could not request results; {e}")
            return


if __name__ == "__main__":
    main()
```
